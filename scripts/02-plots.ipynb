```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
import pandas as pd

arg_df = pd.read_csv("../data/ARG_edu_data.csv")
bra_df = pd.read_csv("../data/BRA_edu_data.csv")
chl_df = pd.read_csv("../data/CHL_edu_data.csv")
ven_df = pd.read_csv("../data/VEN_edu_data.csv")

## ARGENTINA PLOT
# Transpose the relevant data for plotting
arg_years = [int(col[2:]) for col in arg_df.columns if col.startswith('YR')]
arg_primary = arg_df.loc[arg_df['series'] == 'SE.PRM.ENRR', [col for col in arg_df.columns if col.startswith('YR')]].values.flatten()
arg_secondary = arg_df.loc[arg_df['series'] == 'SE.SEC.ENRR', [col for col in arg_df.columns if col.startswith('YR')]].values.flatten()
arg_tertiary = arg_df.loc[arg_df['series'] == 'SE.TER.ENRR', [col for col in arg_df.columns if col.startswith('YR')]].values.flatten()

# Create scatter plot
plt.figure(figsize=(12, 6))

plt.scatter(arg_years, arg_primary, label='Primary Enrollment', color='blue', marker='o')
plt.scatter(arg_years, arg_secondary, label='Secondary Enrollment', color='green', marker='s')
plt.scatter(arg_years, arg_tertiary, label='Tertiary Enrollment', color='red', marker='^')

plt.xlabel('Year')
plt.ylabel('Enrollment (% gross)')
plt.title('Argentina School Enrollment (Primary, Secondary, Tertiary) 2000-2022')
plt.legend()
plt.grid(True)
plt.xticks(arg_years, rotation=45)
plt.tight_layout()
plt.savefig('../figures/arg_enrollment.png')

plt.show()

```
